```python
users = pd.read_csv("Users.csv")
users.head()
```

```python
users.rename(columns={'User-ID':'user_id','Location':'location','Age':'age'},inplace=True)
```

```python
ratings = pd.read_csv("Ratings.csv")
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'},inplace=True)
ratings.head()
```

```python
ratings.shape
```

```python
users.shape
```

```python
books.shape
```

```python
ratings.head()
```

```python
ratings['user_id'].value_counts().shape
```

```python
x=ratings['user_id'].value_counts() > 200
```

```python
y = x[x].index #to get id of users 
y
```

```python
ratings =ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings_with_books = ratings.merge(books,on="ISBN")
ratings_with_books
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'number of rating'},inplace=True)
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating,on='title')
final_rating
```

```python
final_rating=final_rating[final_rating['number of rating']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating
```

```python
books_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
books_pivot.head()
```

```python
books_pivot.shape
```

```python
books_pivot.fillna(0,inplace=True)
```

```python
books_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(books_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model= NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions=model.kneighbors(books_pivot.iloc[8,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
books_pivot.index[237]
```

```python
for i in range(len(suggestions)):
    print(books_pivot.index[suggestions[i]])
```

```python
np.where(books_pivot.index =='Animal Farm')[0][0]
```

```python
def recommend_book(book_name):
    book_id=np.where(books_pivot.index ==book_name)[0][0]
    distances, suggestions=model.kneighbors(books_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    
    for i in range(len(suggestions)):
        if i==0:
            print('The suggestions for',book_name,'are : ')
        if not i:
            print(books_pivot.index[suggestions[i]])
```

```python
recommend_book('Animal Farm')
```

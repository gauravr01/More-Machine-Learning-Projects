```python
df_encoded = pd.get_dummies(X, columns=['team1', 'team2', 'toss_decision', 'toss_winner', 'venue'])
df_encoded.head(30)
```

```python
y1 = pd.get_dummies(Y.winner)
y1
```

```python
y1.head(30)
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
# Create a logistic regression model
model = LogisticRegression()
y_1d = np.argmax(y1.values, axis=1)
# Fit the model using X_encoded as the features and y_encoded as the target variable
model.fit(df_encoded, y_1d)

# Print the trained model's coefficients/weights
print("Coefficients:", model.coef_)

# Print the trained model's intercept/constant term
print("Intercept:", model.intercept_)
```

```python

predictions = model.predict([arr])
team_mapping = {
    0: "Chennai Super Kings",
    1: "Delhi Capitals",
    2: "Kings XI Punjab",
    3 :"Kolkata Knight Riders",
    4: "Mumbai Indians",
    5: "Rajasthan Royals",
    6: "Royal Challengers Bangalore",
    7: "Sunrisers Hyderabad",

}

# Replace numbers with team names in predictions
predictions = [team_mapping[pred] if pred in team_mapping else pred for pred in predictions]

# Print the predictions
print(predictions)
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
get_score(LogisticRegression(), X_train, X_test, y_train, y_test)
```
